```python
# Dependencies
import pandas as pd
```

```python
# load CSV
bootcamp=pd.read_csv("resources/2016-FCC-New-Coders-Survey-Data.csv",low_memory=False)
```

```python
# Read with pandas
bootcamp.head()
```

```python
# Inspect all columns
bootcamp.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
bootcampselect=bootcamp.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
bootcampselect.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
replacements={
    0.0:"No",
    1.0:"Yes"
}
bootcampselect.replace(replacements,inplace=True)
bootcampselect.head()
```

```python
# Calculate total number of respondents in survey
count_survey=len(bootcampselect.index)
```

```python
# Extract rows corresponding only to people who attended a bootcamp
bools=[]
for attendee in bootcampselect['AttendedBootcamp']:
    if attendee == "Yes":
        bools.append(True)
    else:
        bools.append(False)
attendees=pd.Series(bools)
bootcampattendee=bootcampselect[attendees]
count_attendees=len(bootcampselect[attendees].index)
average_age=round(bootcampattendee['Age'].mean(),2)


```

```python
# Calculate average age of attendees

# Calculate how many people attended a bootcamp
bootcamp=round(count_attendees/count_survey*100,2)



# Calculate how many attendees hold degrees
degree=bootcampattendee.groupby('SchoolDegree')
count_degree=degree['AttendedBootcamp'].count().sum()

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification

gender_female=bootcampattendee[bootcampattendee.Gender=="female"]
gender_male=bootcampattendee[bootcampattendee.Gender=="male"]
gender=bootcampattendee["Gender"].count()
print(gender)
gender_female_count=gender_female['AttendedBootcamp'].count()
gender_male_count=gender_male['AttendedBootcamp'].count()
# Calculate percentage of respondents who attended a bootcamp
bootcamp=round(count_attendees/count_survey*100,2)
# Calculate percentage of respondents belonging to each gender
male=round(gender_male_count/gender*100,2)
female=round(gender_female_count/gender*100,2)
# Calculate percentage of attendees with a school degree
percent_degree=round(count_degree/count_attendees*100,2)
# Calculate average post-bootcamp salary

avg_postsal=bootcampattendee["BootcampPostSalary"]
average_post_bootcamp_salary=round(avg_postsal.mean())
```

```python
# Create a new table consolodating above calculations

pdf=pd.DataFrame({"Total Surveyed":count_survey,"Total Bootcamp attendees":count_attendees,"% attended bootcamp":bootcamp, 
           'Aver. Age':average_age,"Has a degree":percent_degree,
          "% Male":male,"% Female":female,"% Non Gender Specific":100-(male+female),"Average Post Bootcamp Salary":average_post_bootcamp_salary
                 },index=[0])

pdf
```

```python
# Improve formatting before outputting spreadsheet
average_post_bootcamp_salary= (f'$ {average_post_bootcamp_salary:,.2f}')
average_post_bootcamp_salary
```
